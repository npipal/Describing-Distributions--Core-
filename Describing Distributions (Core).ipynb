# Answer the following questions:

```python
#Imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
```

```python
df = pd.read_csv('data (4).csv')
df.head()
```

```python
#Features to Analyze
df = df[['VitD_levels', 'Doc_visits', 'TotalCharge']]
df.head()
```

```python
#Creating function for plotting graphs
def plot_function(df, col):
    fig = plt.figure(figsize=(10, 6))
    ax = sns.histplot(x=df[col],kde=True)
    mean = df[col].mean()
    median = df[col].median()
    std = df[col].std()
    ax.axvline(mean, color='red', label=f'Mean = {mean:,.2f}')
    ax.axvline(median, color='green', label=f'Median = {median:,.2f}')
    ax.axvline(mean-std, color = 'black', label = f'─1 StDev = {mean-std:,.2f}')
    ax.axvline(mean+std, color = 'black', label = f'+1 StDev = {mean+std:,.2f}')
    ax.axvspan(mean-std,mean+std,color = 'yellow',zorder = 0)
    ax.set_title(col);
    ax.legend()
```

```python
#VitD_levels
plot_function(df,'VitD_levels')
```

# 'VitD_levels' Answers

1. Is it Discrete or Continuous?
    - Continuous


2. Does it have a skew? If so, which direction (+/-)
    - No Skew


3. What type of kurtosis does it display? (Mesokurtic, Leptokurtic, Platykurtic)
    - Mesokurtic (values cluster in the middle of the distribution - normal)

```python
#Doc_visits Graph
plot_function(df,'Doc_visits')
```

    - Mesokurtic (values cluster in the middle of the distribution - normal)

```python
#TotalCharge Graph
plot_function(df,'TotalCharge')
```
